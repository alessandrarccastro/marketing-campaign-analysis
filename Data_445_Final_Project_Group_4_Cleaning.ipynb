```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dim_country = pd.read_csv('Dim_Country.csv')
dim_customer = pd.read_csv('Dim_Customer.csv')
dim_education = pd.read_csv('Dim_Education.csv')
dim_marital_status = pd.read_csv('Dim_Marital_Status.csv')
fct_campaign_impact = pd.read_csv('Fct_Campaign_Impact.csv')
fct_customer_behavior = pd.read_csv('Fct_Customer_Behavior.csv')
fct_customer_demographic = pd.read_csv('Fct_Customer_Demographic.csv')
```

```python
# print heads
print(dim_country.head())
print(dim_customer.head())
print(dim_education.head())
print(dim_marital_status.head())
print(fct_campaign_impact.head())
print(fct_customer_behavior.head())
print(fct_customer_demographic.head())
```

```python
# show all data types
print(dim_country.dtypes)
print(dim_customer.dtypes)
print(dim_education.dtypes)
print(dim_marital_status.dtypes)
print(fct_campaign_impact.dtypes)
print(fct_customer_behavior.dtypes)
print(fct_customer_demographic.dtypes)
```

```python
# correct data types
# column Income - fct_customer_demographic

# correct column name ={' Income ':'Income'}
fct_customer_demographic = fct_customer_demographic.rename(columns={' Income ':'Income'})

# # remove $ and ,
fct_customer_demographic['Income'] = fct_customer_demographic['Income'].str.replace('$', '')
fct_customer_demographic['Income'] = fct_customer_demographic['Income'].str.replace(',', '')
fct_customer_demographic['Income'] = fct_customer_demographic['Income'].astype(str).str.replace(' ', '')
fct_customer_demographic['Income'] = fct_customer_demographic['Income'].astype('float32')

fct_customer_demographic.head()
```

```python
# print describe for all datasets
print(dim_country.describe())
print(dim_customer.describe())
print(dim_education.describe())
print(dim_marital_status.describe())
print(fct_campaign_impact.describe())
print(fct_customer_behavior.describe())
print(fct_customer_demographic.describe())
```

```python
# drop age > 100  Fct_Customer_Demographic
fct_customer_demographic = fct_customer_demographic[fct_customer_demographic['Age'] < 100]

# sort by age
fct_customer_demographic_sorted = fct_customer_demographic.sort_values(by=['Age'])
print(fct_customer_demographic_sorted['Age'].tail())

```

```python
# print (missing values)
print(dim_country.isnull().sum())
print(dim_customer.isnull().sum())
print(dim_education.isnull().sum())
print(dim_marital_status.isnull().sum())
print(fct_campaign_impact.isnull().sum())
print(fct_customer_behavior.isnull().sum())
print(fct_customer_demographic.isnull().sum())
```

```python
# use median to fill missing values in Income column

fct_customer_demographic['Income'] = fct_customer_demographic['Income'].fillna(fct_customer_demographic['Income'].median())
print(fct_customer_demographic.isnull().sum())
```

```python
# check duplicates on Customer_ID
print(fct_customer_demographic.duplicated(subset=['Customer_ID']).sum())
```
