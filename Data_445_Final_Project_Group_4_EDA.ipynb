```python
# Identify boolean columns
boolean_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'Complain']

# Sum the boolean columns
boolean_sums = data[boolean_columns].sum()

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=boolean_sums.index, y=boolean_sums.values)
plt.xlabel('Columns')
plt.ylabel('Sum of 0 and 1 values')
plt.title('Sum of Boolean Columns (0 and 1)')
plt.show()
```
